```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customers.csv")
print(df.head())
```

Bar Chart using Matplotlib

```python
# Graph 1: Country-wise Customers Count
country_count = df['country'].value_counts()

plt.figure()
plt.bar(country_count.index, country_count.values)
plt.xlabel("Country")
plt.ylabel("Number of Customers")
plt.title("Customers per Country")
plt.xticks(rotation=75)
plt.show()
```

Histogram (of numerical data) Matplotlib

```python
#Graph 2: Credit Limit Distribution
plt.figure()
plt.hist(df['creditLimit'], bins=5)
plt.xlabel("Credit Limit")
plt.ylabel("Frequency")
plt.title("Credit Limit Distribution")
plt.show()
```

Seaborn Satistical(Box) plot

```python
#Graph 3: Credit Limit by Country (Boxplot)
plt.figure()
sns.boxplot(x='country', y='creditLimit', data=df)
plt.xticks(rotation=75)
plt.title("Credit Limit by Country")
plt.show()
```

Seaborn Countplot

```python
#Graph 4: Customers per Country (Countplot)
plt.figure()
sns.countplot(x='country', data=df)
plt.xticks(rotation=75)
plt.title("Number of Customers per Country")
plt.show()
```

Seaborn + Matplotlib (Bar plot)

```python
#Graph 5: Seaborn plot + Matplotlib labels
plt.figure()
sns.barplot(x='country', y='creditLimit', data=df)
plt.xlabel("Country")
plt.ylabel("Average Credit Limit")
plt.title("Average Credit Limit by Country")
plt.xticks(rotation=75)
plt.show()
```

Seaborn + Matplotlib (Count plot)

```python
#Graph 6: Seaborn style + Matplotlib plot
sns.set_style("whitegrid")

country_count.plot(kind='bar')
plt.xlabel("Country")
plt.ylabel("Customers")
plt.title("Customers by Country (Mixed)")
plt.show()
```
