```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("products.csv")
print(df.head())
```

Seaborn + Matplotlib

GRAPH 1: Violin Plot
âžœ Product Line vs Buy Price (distribution + density)

Violin plot shows: distribution,density,spread of values

```python
plt.figure(figsize=(10,6))
sns.violinplot(
    x="productLine",
    y="buyPrice",
    data=df,
    palette="viridis"
)
plt.title("Buy Price Distribution by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Buy Price")
plt.show()
```

GRAPH 2: Scatter Plot with Color Encoding
âžœ Buy Price vs MSRP (Price comparison)

Scatter Shows: relationship between two prices, color shows product category

```python
plt.figure(figsize=(10,5))
sns.scatterplot(
    x="buyPrice",
    y="MSRP",
    hue="productLine",
    data=df,
    s=100
)
plt.title("Buy Price vs MSRP")
plt.xlabel("Buy Price")
plt.ylabel("MSRP")
plt.show()
```

GRAPH 3: Heatmap (Correlation)
âžœ Numerical feature relationships

Heatmap shows: strong / weak relationships, very common in ML & AI analysis

```python
plt.figure(figsize=(10,6))
corr = df[['quantityInStock','buyPrice','MSRP']].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```

GRAPH 4: Line Plot (Trend-style Visualization)
âžœ Quantity vs MSRP (Sorted view)

Line shows: Trend visualization, Shows how stock varies with price

```python
df_sorted = df.sort_values("MSRP")

plt.figure(figsize=(16,8))
sns.lineplot(
    x="MSRP",
    y="quantityInStock",
    data=df_sorted,
    marker="o",
    color="purple"
)
plt.title("Stock Quantity vs MSRP Trend")
plt.xlabel("MSRP")
plt.ylabel("Quantity in Stock")
plt.show()
```
